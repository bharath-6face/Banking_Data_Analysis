# Handling the columns wise missing value

```python
df['Gender'].mode()
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
```

```python
df['Gender'].isnull().sum()
```

```python
df['Age'].median()
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df["Age"].isnull().sum()
```

```python
df['Occupation'].mode()
```

```python
df['Occupation'] = df['Occupation'].fillna(df['Occupation'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df['Monthly_Income'] = df['Monthly_Income'].fillna(df['Monthly_Income'].median())
df['Account_Type'] = df['Account_Type'].fillna(df['Account_Type'].mode()[0])
df['Transaction_Type'] = df['Transaction_Type'].fillna(df['Transaction_Type'].mode()[0])
df['Account_Status'] = df['Account_Status'].fillna(df['Account_Status'].mode()[0])
df['Transaction_Amount'] = df['Transaction_Amount'].fillna(df['Transaction_Amount'].median())
df['Account_Balance'] = df['Account_Balance'].fillna(df['Account_Balance'].median())
```

```python
df.isnull().sum()
```

```python
df["Customer_Segment"] =df["Customer_Segment"].fillna(df["Customer_Segment"].mode()[0])
```

```python
df['Relationship_Years'] =df['Relationship_Years'].fillna(df['Relationship_Years'].median())
```

```python
df['Loan_Customer'] = df['Loan_Customer'].fillna(df['Loan_Customer'].mode()[0])
df['Transaction_Channel'] = df['Transaction_Channel'].fillna(df['Transaction_Channel'].mode()[0])
df['Payment_Method'] = df['Payment_Method'].fillna(df['Payment_Method'].mode()[0])
df['Merchant_Category'] = df['Merchant_Category'].fillna(df['Merchant_Category'].mode()[0])
df['Currency'] = df['Currency'].fillna(df['Currency'].mode()[0])
df['Branch'] = df['Branch'].fillna(df['Branch'].mode()[0])
df['Transaction_Date'] = df['Transaction_Date'].fillna(df['Transaction_Date'].mode()[0])
df['Transaction_Time'] = df['Transaction_Time'].fillna(df['Transaction_Time'].mode()[0])
df['Day_of_Week'] = df['Day_of_Week'].fillna(df['Day_of_Week'].mode()[0])
df['Device_Type'] = df['Device_Type'].fillna(df['Device_Type'].mode()[0])
df['Login_Location'] = df['Login_Location'].fillna(df['Login_Location'].mode()[0])
```

```python
df['Transaction_Status'] = df['Transaction_Status'].fillna(df['Transaction_Status'].mode()[0])
df['Fraud_Flag'] = df['Fraud_Flag'].fillna(df['Fraud_Flag'].mode()[0])
```

```python
df['Credit_Score'] = df['Credit_Score'].fillna(df['Credit_Score'].median())
df['Remarks'] = df['Remarks'].fillna(df['Remarks'].mode()[0])
```

```python
df.isnull().sum()
```

Drop for null value in IFSC_Code column

```python
df = df.dropna(subset=['IFSC_Code'])
```
